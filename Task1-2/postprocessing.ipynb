## Completed form preprocessing

The form has 3 worksheets for Day 1, 2 and 3. Each worksheet will be considered as a separate dataframe and then merged into one single dataframe to export to csv.

```python
df1 = pd.read_excel('completed_form.xlsx', 'Day 1', engine='openpyxl')
df1_date = df1.iloc[1,1]
df2 = pd.read_excel('completed_form.xlsx', 'Day 2', engine='openpyxl')
df2_date = df2.iloc[1,1]
df3 = pd.read_excel('completed_form.xlsx', 'Day 3', engine='openpyxl')
df3_date = df3.iloc[1,1]
```

```python
# remove border and drop rows with no data
df1 = df1.iloc[4:100, 1:].dropna(axis=0, how='all')
df2 = df2.iloc[4:100, 1:].dropna(axis=0, how='all')
df3 = df3.iloc[4:100, 1:].dropna(axis=0, how='all')
```

```python
# 'o' mark in the cells to 1
df1.iloc[:, 2:23] = df1.iloc[:, 2:23].replace({'o':int(1)})
df2.iloc[:, 2:23] = df2.iloc[:, 2:23].replace({'o':int(1)})
df3.iloc[:, 2:23] = df3.iloc[:, 2:23].replace({'o':int(1)})
```

```python
# date format to %d/%m/%y
df1['StartDate'] = df1_date
df1['StartDate'] = pd.to_datetime(df1['StartDate']).dt.strftime('%d/%m/%y')
df2['StartDate'] = df2_date
df2['StartDate'] = pd.to_datetime(df2['StartDate']).dt.strftime('%d/%m/%y')
df3['StartDate'] = df3_date
df3['StartDate'] = pd.to_datetime(df3['StartDate']).dt.strftime('%d/%m/%y')
```

```python
# merge df1, 2 and 3
frames = [df1, df2, df3]
df = pd.concat(frames)
df
```

```python
# PrimaryActivityCode and SecondaryActivityCode
activity_list = df.iloc[:, 2:13].fillna(0).to_numpy()

def ActivityCodeSorter(array):
    activity_code = ('ED145', 'EL642', 'X893', 'H179', 'L418', 'R523', 'S801',
           'T695', 'UD415', 'PW101', 'CW982', 'O733')
    primary = []
    secondary = []
    for row in array:
        if any(row==2):
            primary.append(activity_code[np.argmax(row==1)])
            secondary.append(activity_code[np.argmax(row==2)])
        else:
            primary.append(activity_code[np.argmax(row==1)])
            secondary.append(np.nan)
    return primary, secondary

PrimaryActivityCode, SecondaryActivityCode = ActivityCodeSorter(activity_list)
print(PrimaryActivityCode)
print(SecondaryActivityCode)
```

```python
# DescriptiveLabel
DescriptiveLabel = df.iloc[:, 14].to_numpy()
print(DescriptiveLabel)
```

```python
# StartDate
StartDate = df['StartDate'].to_numpy()
print(StartDate)
df['StartDate'] = pd.to_datetime(df['StartDate'], format='%d/%m/%y')
```

```python
# UniBathWeekNo and DayOfWeek
days_list = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

df['UniBathWeekNo'] = df['StartDate'].apply(lambda x: x.isocalendar()[1]+14)
df['DayOfWeek'] = df['StartDate'].apply(lambda x: days_list[x.isocalendar()[2]-1])
```

```python
UniBathWeekNo = df['UniBathWeekNo'].to_numpy()
DayOfWeek = df['DayOfWeek'].to_numpy()
```

```python
# StartTime
df['StartTime'] = pd.to_datetime(df.iloc[:,0], format='%H:%M:%S').dt.time
StartTime = df['StartTime'].to_numpy(dtype=str)
StartTime = [i[:-3] for i in StartTime]
print(StartTime)
```

```python
# EndTime
df['EndTime'] = pd.to_datetime(df.iloc[:,1], format='%H:%M:%S').dt.time
EndTime = df['EndTime'].to_numpy(dtype=str)
EndTime = [i[:-3] for i in EndTime]
print(EndTime)
```

```python
# DurationMins
from datetime import datetime, date

DurationMins = []

today = date(1, 1, 1)
tomorrow = date(1, 1, 2)

for i, start in enumerate(df['StartTime']):
    
    if df['EndTime'].iloc[i] < start:
        delta = datetime.combine(tomorrow, df['EndTime'].iloc[i]) - datetime.combine(today, start)
        
    else:
        delta = datetime.combine(today, df['EndTime'].iloc[i]) - datetime.combine(today, start)
    
    DurationMins.append(int(delta.total_seconds() / 60))
    
print(DurationMins)
```

```python
# EnjoymentScore

enjoyment_list = df.iloc[:, 15:20].fillna(0).to_numpy()

def EnjoymentScoreSorter(array):
    enjoyment_score = (-2, -1, 0, 1, 2)
    scores = []
    for row in array:
        if all(row==0):
            scores.append(np.nan)
        else:
            scores.append(enjoyment_score[np.argmax(row)])

    return scores

EnjoymentScore = EnjoymentScoreSorter(enjoyment_list)
print(EnjoymentScore)
```

```python
# AdditionalData1
Sleepiness = []

sleepiness_list = df.iloc[:, 20:23].fillna(0).to_numpy()

def SleepinessScoreSorter(array):
    sleepiness_score = (-1, 0, 1)
    scores = []
    for row in array:
        if all(row==0):
            scores.append(np.nan)
        else:
            scores.append(sleepiness_score[np.argmax(row)])

    return scores

SleepinessScore = SleepinessScoreSorter(sleepiness_list)
print(SleepinessScore)
```

```python
# AdditionalData2
SocialMediaMinutes = df.iloc[:, 23].fillna(0).to_numpy()
print(SocialMediaMinutes)
```

```python
# make lists into one dataframe for export
df_csv_data = np.array([PrimaryActivityCode, SecondaryActivityCode, 
               DescriptiveLabel, StartDate, UniBathWeekNo,
               DayOfWeek, StartTime, EndTime, DurationMins, 
               EnjoymentScore, SleepinessScore, SocialMediaMinutes
              ]).T
```

```python
df_csv = pd.DataFrame(df_csv_data, columns=['PrimaryActivityCode', 'SecondaryActivityCode', 
               'DescriptiveLabel', 'StartDate', 'UniBathWeekNo',
               'DayOfWeek', 'StartTime', 'EndTime', 'DurationMins', 
               'EnjoymentScore', 'SleepinessScore', 'SocialMediaMinutes'
              ])
```
